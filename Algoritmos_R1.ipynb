## Torres de Hanoi con Divide y vencerás

Resuelve el problema de las [Torres de Hanoi](https://es.wikipedia.org/wiki/Torres_de_Han%C3%B3i) para un número arbitrario de fichas utilizando la técnica de divide y vencerás. 

```python
# Torres de Hanoi

def torres_hanoi(num_fichas, origen, auxiliar, destino):
    # De ser una ficha, se mueve directamente de origen a destino
    if num_fichas == 1:
        # Indicamos el movimiento que realizamos
        print('Ficha:', num_fichas, '\tMovimiento:', origen , ' -> ', destino)
        
    else:
        torres_hanoi (num_fichas - 1 , origen , destino , auxiliar)
        # Indicamos el movimiento que realizamos
        print('Ficha:', num_fichas, '\tMovimiento:', origen, ' -> ', destino)
        torres_hanoi (num_fichas - 1, auxiliar, origen, destino)
        
# T1 -> Torre de origen, T2->Torre auxiliar o intermedia, T3->Torre de destino de las fichas
num_fichas = 4
print('Realizando algoritmo con', num_fichas, 'fichas')
torres_hanoi(num_fichas, 'T1', 'T2', 'T3')

num_fichas = 7
print('\nRealizando algoritmo con', num_fichas, 'fichas')
torres_hanoi(num_fichas, 'T1', 'T2', 'T3')
```

## Sucesión de Fibonacci

Cálcula el n-ésimo término de la [Sucesión de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci) mediante un algoritmo recursivo y otro iterativo. Representa gráficamente cómo crece el tiempo de cómputo en función del número de términos para ambos algoritmos. 

```python
#Sucesión_de_Fibonacci
import matplotlib.pyplot as plt
import time

# Función que calcula el enésimo término de Fibonacci de forma recursiva
def fibonacci_recursivo(enesimo_termino):
    # Si el enésimo término es 0 o 1, se devuelve el propio valor
    if enesimo_termino < 2:
        return enesimo_termino

    # De no serlo, se calculan los valores previos de la serie y se añaden al total calculado 
    return fibonacci_recursivo(enesimo_termino - 1) + fibonacci_recursivo(enesimo_termino - 2)


# Función que calcula el enésimo término de Fibonacci de forma iterativa
def fibonacci_iterativo(enesimo_termino):
    a = 0
    b = 1

    i = 0
    while i < enesimo_termino:
        #Ejecutamos las operaciones iterativamente hasta la obtención del valor de Fibonacci para el término enésimo, empezando por los primeros términos de la serie
        b = b + a
        a = b - a
        i = i + 1

    return a

# Indicamos los términos enésimos que se calcularán
enesimos_terminos = [13, 24]

# Inicializamos las listas que contendrán los tiempos de ejecución
tiempos_recursivo = []
tiempos_iterativo = []

#Obtención de los enésimos términos de Fibonacci mediante el algoritmo recursivo
print('Algoritmo recursivo')

i = 0
while i < len(enesimos_terminos):
    # Obtenemos la marca de tiempo de inicio
    comienzo_recursivo = time.time_ns()

    # Llamamos a la función pasando el valor como entero y eliminando posibles espacios que tenga
    print('Término:', enesimos_terminos[i], '\tValor:', fibonacci_recursivo(enesimos_terminos[i]))

    # Obtenemos la marca de tiempo de finalización                         
    fin_recursivo = time.time_ns()
    tiempos_recursivo.append((fin_recursivo - comienzo_recursivo) / 1000000)
    i = i + 1

# Obtención de los enésimos términos de Fibonacci mediante el algoritmo recursivo
print('\nAlgoritmo iterativo')

i = 0
while i < len(enesimos_terminos):
    # Obtenemos la marca de tiempo de inicio
    comienzo_iterativo = time.time_ns()

    # Llamamos a la función pasando como parámetro el valor del enésimo término
    print('Término:', enesimos_terminos[i], '\tValor:', fibonacci_iterativo(enesimos_terminos[i]))

    # Obtenemos la marca de tiempo de finalización                         
    fin_iterativo = time.time_ns()
    tiempos_iterativo.append((fin_iterativo - comienzo_iterativo) / 1000000)
    i = i + 1

# Procedemos a crear la gráfica que nos permitirá mostrar los tiempos de ejecución
fig, ax = plt.subplots()

# Configuramos los valores de los ejes X e y para los resultados de ambos métodos
ax.plot(enesimos_terminos, tiempos_recursivo, linewidth=2.0, label='Método recursivo', color='red')
ax.plot(enesimos_terminos, tiempos_iterativo, linewidth=2.0, label='Método iterativo', color='green')
ax.set_xlabel('Términos')
ax.set_ylabel('Tiempo (ms)')
ax.legend()

# Mostramos la gráfica
plt.show()
```

## Devolución de cambio por técnica voraz

Resuelve el [Problema del Cambio de Moneda](https://es.wikipedia.org/wiki/Problema_de_cambio_de_monedas) utilizando una técnica voraz.

```python
# Problema del cambio de moneda

def cambio_monedas(monedas, cant_a_cambiar):
    # Nos aseguramos que las monedas están ordenadas de mayor a menor
    monedas.sort(reverse=True)
    print('Cantidad a cambiar:', cant_a_cambiar)
    
    total_cambio = 0
    lista_cambio = []
    i = 0
    while cant_a_cambiar > 0:
        # Buscamos la moneda de mayor valor que no supere la cantidad a cambiar
        while i < len(monedas) and monedas[i] > cant_a_cambiar:
            i = i + 1
        
        # Dividimos la cantidad a cambiar por la moneda de máximo valor posible
        cant_monedas = int(cant_a_cambiar / monedas[i])
        cant_a_cambiar = cant_a_cambiar % monedas[i]

        lista_cambio.append(str(cant_monedas) + 'x' + str(monedas[i]))

    return lista_cambio

# Indicamos las monedas de las que se dispone
monedas = [50, 20, 10, 5, 2, 1]

print('Monedas disponibles:', ', '.join(str(moneda) for moneda in monedas),'\n')
lista_cambio = cambio_monedas(monedas, 363)
print('Cambio: ' + ' + '.join(lista_cambio) + '\n')

lista_cambio = cambio_monedas(monedas, 43)
print('Cambio: ' + ' + '.join(lista_cambio))
```

## N-Reinas por técnica de vuelta atrás
Resuelve el [Problema de las N-Reinas](https://es.wikipedia.org/wiki/Problema_de_las_ocho_reinas) en un tablero de dimensión N mediante la técnica de la vuelta atrás (backtraking).

```python
# Problema de las N-Reinas

import numpy as np
import time
from tabulate import tabulate

# Esta función comprueba si ya hay una reina en cualquiera de las diagonales, empezando desde la posición donde se quiere colocar la nueva. En cuanto se encuentra una se devuelve True y se sale de la función
def comp_reinas_diagonales(tablero, i, j):
    dim_tablero = tablero.shape[0]

    # Recorrido hacia arriba e izquierda
    i_aux = i
    j_aux = j
    while i_aux > 0 and j_aux > 0:
        i_aux = i_aux - 1
        j_aux = j_aux - 1

        if tablero[i_aux, j_aux] == 1:
            return True

    # Recorrido hacia abajo y derecha
    i_aux = i
    j_aux = j
    while i_aux < dim_tablero - 1 and j_aux < dim_tablero - 1:
        i_aux = i_aux + 1
        j_aux = j_aux + 1

        if tablero[i_aux, j_aux] == 1:
            return True

    # Recorrido hacia abajo e izquierda
    i_aux = i
    j_aux = j
    while i_aux < dim_tablero - 1 and j_aux > 0:
        i_aux = i_aux + 1
        j_aux = j_aux - 1

        if tablero[i_aux, j_aux] == 1:
            return True

    # Recorrido hacia arriba y derecha
    i_aux = i
    j_aux = j
    while i_aux > 0 and j_aux < dim_tablero - 1:
        i_aux = i_aux - 1
        j_aux = j_aux + 1

        if tablero[i_aux, j_aux] == 1:
            return True

    # Si llega hasta aquí es que no hay reinas en las diagonales
    return False

# Función que va colocando las reinas en el tablero
def posicionar_reinas_en_tablero(dim_lado_tablero):
    print('Número de reinas a colocar:', dim_lado_tablero)
    print('Fecha de comienzo:', time.strftime('%d/%m/%Y %H:%M:%S.%m'))
    
    # Creamos matriz vacía nxn, siendo n = dim_lado_tablero
    tablero = np.zeros([dim_lado_tablero, dim_lado_tablero], dtype=int)
    
    i = 0
    while i < dim_lado_tablero:
        j = 0
    
        while j < dim_lado_tablero:
            if tablero[i, j] != 1 and 1 not in tablero[i, :] and 1 not in tablero[:, j] and comp_reinas_diagonales(tablero, i, j) is False:
                tablero[i, j] = 1
                break
    
            else:
                j = j + 1
    
        # Si no se pudo colocar ninguna reina en la fila gestionamos en función de la fila en que nos encontremos
        if 1 not in tablero[i, :]:
            # Buscamos si hay alguna fila donde la reina no esté colocada a la derecha del todo
            i = i - 1
            while tablero[i, :][-1] == 1 and i > 0:
                i = i - 1
    
            # Esto no debería suceder. En caso de que suceda, terminamos el programa para evitar un bucle infinito
            if tablero[0, :][-1] == 1:
                print('No he podido resolver el problema. Revisa el código')
                exit(1)
    
            # Ya localizada la fila, desplazamos a la reina una posición a la derecha, borramos las filas que queden por debajo y seguimos desde ahí
            tablero[i, :] = np.roll(tablero[i, :], 1)
            tablero[i+1:, :] = 0
    
        else:
            i = i + 1
    
    print('Fecha de finalización:', time.strftime('%d/%m/%Y %H:%M:%S.%m'))
    print(tabulate(tablero, tablefmt='grid'), '\n')

# La dimensión del lado del tablero equivale al número de reinas a colocar
num_reinas = 8
posicionar_reinas_en_tablero(num_reinas)

num_reinas = 32
posicionar_reinas_en_tablero(num_reinas)
```
